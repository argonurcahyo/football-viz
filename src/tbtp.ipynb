```python
import pandas as pd
```

```python
btp = pd.read_excel('../data/btp.xlsx')
```

```python
btp['no_eks'] = btp['no_eks'].astype('int')
btp['flag_aktif'] = btp['flag_aktif'].astype('int')
btp[:10]
```

```python
txt = []
data = btp.copy()
for i in range(len(data)):
    txt.append(
        f"('{data['id_publikasi'].iloc[i]}',{data['no_eks'].iloc[i]},'{data['box_num'].iloc[i]}')")
```

```python
page = 1
for page in range(1,8):
    txt2 = ",".join(txt[(page-1)*1000:page*1000])
    txt_file = open(f'../data/btp_{page}.txt','w')
    txt_file.write(txt2)
    txt_file.close()
```
