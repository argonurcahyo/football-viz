```python
import pandas as pd
```

```python
tbtp = pd.read_excel('../data/tbtp.xlsx')
```

```python
tbtp[tbtp['box_num']==1]
```

```python
txt = []
data = tbtp.copy()
for i in range(len(data)):
    txt.append(
        f"('{data['id_publikasi'].iloc[i]}',{data['no_eks'].iloc[i]},'{data['box_num'].iloc[i]}')")
```

```python
page = 1
for page in range(1,18):
    txt2 = ",".join(txt[(page-1)*1000:page*1000])
    txt_file = open(f'../data/tbtp_{page}.txt','w')
    txt_file.write(txt2)
    txt_file.close()
```
