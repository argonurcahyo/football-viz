```python
def get_scenes_by_page(driver, page=1):
    zz_url = f'https://www.brazzers.com/videos/page/{page}'
    zz_scenes_details = pd.DataFrame()
    data = []

    driver.get(zz_url)
    time.sleep(10)
    scenediv = driver.find_elements(
        By.CSS_SELECTOR, '.e1vusg2z1 .e1jyqorn29')

    for i in range(len(scenediv)):
        # hover thumbnail
        action = ActionChains(driver)
        hover_thumb = scenediv[i].find_element(
            By.CSS_SELECTOR, '.eh7jk9w0')
        action.move_to_element(hover_thumb).perform()
        time.sleep(1)

        try:
            preview_url = hover_thumb.find_element(
                By.CSS_SELECTOR, 'video').get_attribute('src')
        except:
            preview_url = None
        try:
            scene_title = scenediv[i].find_element(
                By.CSS_SELECTOR, '.e1qkfw3j0 a').get_attribute('title')
        except:
            scene_title = None
        try:
            scene_url = scenediv[i].find_element(
                By.CSS_SELECTOR, '.e1qkfw3j0 a').get_attribute('href')
        except:
            scene_url = None
        try:
            scene_id = scene_url.split('/')[4]
        except:
            scene_id = None
        try:
            scene_img = scenediv[i].find_element(
                By.CSS_SELECTOR, 'a picture img').get_attribute('src')
        except:
            scene_img = None
        try:
            scene_performer = scenediv[i].find_element(
                By.CSS_SELECTOR, '.e1jyqorn17 .e1jyqorn11').text
        except:
            scene_performer = None
        try:
            scene_date = scenediv[i].find_element(
                By.CSS_SELECTOR, '.e1jyqorn17 .e1jyqorn24').text
        except:
            scene_date = None

        print(f'{page}|{i+1:02}. {scene_title}')

        scene_detail = {
            'id': scene_id,
            'title': scene_title,
            'performer': scene_performer,
            'release_date': scene_date,
            'url': scene_url,
            'img_url': scene_img,
            'preview_url': preview_url
        }
        data.append(scene_detail)

    zz_scenes_details = pd.DataFrame(data)
    return zz_scenes_details
```

```python
scenes_df = pd.DataFrame()
for page in range(1, 11):
    d = get_scenes_by_page(driver, page)
    scenes_df = pd.concat([scenes_df, d]).reset_index(drop=True)
```

```python
scenes_df.to_csv('../../data/zz10.csv')
```
