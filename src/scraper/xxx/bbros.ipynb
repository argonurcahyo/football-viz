```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains

from webdriver_manager.chrome import ChromeDriverManager

import pandas as pd
import time
import json

options = Options()
options.headless = True

driver = webdriver.Chrome(options=options,service=ChromeService(ChromeDriverManager().install()))
```

```python
def scrape_bbros(driver, page=1):
    bbros_url = f'https://bangbros.com/videos/{page}'
    driver.get(bbros_url)

    th = driver.find_elements(By.CSS_SELECTOR, '.echThumb')

    lis = []
    for i in range(len(th)):
        models = th[i].find_elements(By.CSS_SELECTOR, '.cast-wrapper .cast')

        try:
            id = th[i].find_element(
                By.CSS_SELECTOR, 'a.thmb_lnk').get_attribute('id')
        except:
            id = None
        try:
            title = th[i].find_element(
                By.CSS_SELECTOR, 'a.thmb_lnk').get_attribute('title')
        except:
            title = None
        try:
            url = th[i].find_element(
                By.CSS_SELECTOR, 'a.thmb_lnk').get_attribute('href')
        except:
            url = None
        try:
            site = th[i].find_element(By.CSS_SELECTOR, '.thmb_mr_1').text
        except:
            site = None
        try:
            release_date = th[i].find_element(
                By.CSS_SELECTOR, '.thmb_mr_2').text
        except:
            release_date = None
        try:
            img_url = th[i].find_element(
                By.CSS_SELECTOR, 'img.thmbPic_tag').get_attribute('data-src')
        except:
            img_url = None
        try:
            duration = th[i].find_element(By.CSS_SELECTOR, 'b.tTm').text
        except:
            duration = None

        actors = []
        for k in range(len(models)):
            try:
                actors.append(models[k].text.replace('\n', ''))
            except:
                continue
        actors = ", ".join(actors)

        print(f"({page}|{i+1}) {title}")

        lis.append({
            'id': id,
            'title': title,
            'actors': actors,
            'site': site,
            'release_date': release_date,
            'url': url,
            'img_url': img_url,
            'duration': duration
        })
    return pd.DataFrame(lis)
```

```python
df = pd.DataFrame()
for page in range(1,295):
    d = scrape_bbros(driver,page)
    df = pd.concat([df, d]).reset_index(drop=True)
```
