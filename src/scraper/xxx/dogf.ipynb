```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains

from webdriver_manager.chrome import ChromeDriverManager

import pandas as pd
import time
import json

options = Options()
options.headless = True

driver = webdriver.Chrome(options=options,service=ChromeService(ChromeDriverManager().install()))
```

```python
dogf_url = 'https://www.dogfartnetwork.com/tour/scenes/?p=2&nats=MC4wLjIzLjIzLjAuMC4wLjAuMA'
```

```python
driver.get(dogf_url)

th = driver.find_elements(By.CSS_SELECTOR, '.recent-updates')
```

```python
img_url = th[0].find_element(By.CSS_SELECTOR,'.container-9s-preview img').get_attribute('src')
url = th[0].find_element(By.CSS_SELECTOR,'a.thumbnail').get_attribute('href')
actors = th[0].find_element(By.CSS_SELECTOR,'.recent-details-title h3.scene-title').text
site = th[0].find_element(By.CSS_SELECTOR,'.recent-details-title p.help-block').text
```

```python
def scrape_dogf(driver, page=1):
    url = f'https://www.dogfartnetwork.com/tour/scenes/?p={page}&nats=MC4wLjIzLjIzLjAuMC4wLjAuMA'

    driver.get(url)

    th = driver.find_elements(By.CSS_SELECTOR, '.recent-updates')

    lis = []
    for i in range(len(th)):
        try:
            img_url = th[i].find_element(
                By.CSS_SELECTOR, '.container-9s-preview img').get_attribute('src')
        except:
            img_url: None
        try:
            url = th[i].find_element(
                By.CSS_SELECTOR, 'a.thumbnail').get_attribute('href')
        except:
            url = None
        try:
            actors = th[i].find_element(
                By.CSS_SELECTOR, '.recent-details-title h3.scene-title').text
        except:
            actors = None
        try:
            site = th[i].find_element(
                By.CSS_SELECTOR, '.recent-details-title p.help-block').text
        except:
            site = None
        print(f'({page}|{i+1}) {actors} in {site}')
        lis.append({
            'actors': actors,
            'site': site,
            'url': url,
            'img_url': img_url,
            'have': False
        })
    return pd.DataFrame(lis)
```

```python
df = pd.DataFrame()
dfw = pd.DataFrame()
count = 1
for i in range(1,135):
    d = scrape_dogf(driver,i)
    df = pd.concat([df,d]).reset_index(drop=True)
    dfw = pd.concat([dfw,d]).reset_index(drop=True)
    
    if i%10==0:        
        dfw.to_excel(f'../../data/dogf_{count}.xlsx')
        count+=1
        dfw = pd.DataFrame()
```
