```python
import pandas as pd
```

```python
wc2006 = pd.read_excel('../../../data/WC.xlsx',
                       sheet_name='WC2006', index_col=0)
wc2010 = pd.read_excel('../../../data/WC.xlsx',
                       sheet_name='WC2010', index_col=0)
wc2014 = pd.read_excel('../../../data/WC.xlsx',
                       sheet_name='WC2014', index_col=0)
wc2018 = pd.read_excel('../../../data/WC.xlsx',
                       sheet_name='WC2018', index_col=0)
wc2022 = pd.read_excel('../../../data/WC.xlsx',
                       sheet_name='WC2022', index_col=0)
```

```python
wc = wc2022[['country', 'confederation', 'group_2022']].merge(
    wc2018[['country', 'confederation', 'group_2018']], on=['country', 'confederation'], how='outer').merge(
        wc2014[['country', 'confederation', 'group_2014']], on=['country', 'confederation'], how='outer').merge(
            wc2010[['country', 'confederation', 'group_2010']], on=['country', 'confederation'], how='outer').merge(
                wc2006[['country', 'confederation', 'group_2006']], on=['country', 'confederation'], how='outer'
)
```

```python
wc[wc['confederation']=='AFC'].fillna('-')
```

```python
wc['count']=[wc[['group_2022','group_2018','group_2014','group_2010','group_2006']].iloc[i].count() for i in range(len(wc))]
```

```python
wc.sort_values(['count','confederation'],ascending=[False,True])
```
