```python
pus_subyek = pd.read_csv('../../../data/perpus_subyek.csv',index_col=0)
portal_kat = pd.read_excel('../../../data/portal_pub_kat.xlsx',index_col=0)
```

```python
merged = portal_kat.merge(pus_subyek, on='kd_subyek',
                          how='left').reset_index(drop=True)
merged_nan = merged[merged['subyek'].isna()].reset_index(drop=True)
merged_nan['katalog'] = merged_nan['katalog'].map(lambda x: x.replace(
    "'", "").replace('"', '').replace('”', '').replace('“', '').replace('`', ''))
```

```python
pus_wilda = pd.read_csv('../../../data/t_history_wilda.csv',
                        index_col=0, converters={'kd_prop': str, 'kd_kab': str, 'kd_kec': str})
portal_wilda = pd.read_excel('../../../data/portal_wilda.xlsx', index_col=0,
                             converters={'kd_prop': str, 'kd_kab': str, 'kd_kec': str})
```

```python
wilda_merge = portal_wilda.merge(
    pus_wilda, on=['kd_prop', 'kd_kab', 'kd_kec'], how='left')
wilda_merge_nan = wilda_merge[wilda_merge['nm_wilda_y'].isna()].reset_index(drop=True)
wilda_merge_nan['nm_wilda_x'] = wilda_merge_nan['nm_wilda_x'].map(lambda x:x.replace('Kecamatan ','').replace('Distrik ',''))
```
