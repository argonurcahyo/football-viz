```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from webdriver_manager.chrome import ChromeDriverManager

import pandas as pd
import time

options = Options()
options.headless = False

driver = webdriver.Chrome(options=options,service=ChromeService(ChromeDriverManager().install()))
```

```python
amazing_spiderman_2022_url = 'https://www.marvel.com/comics/series/32866/the_amazing_spider-man_2022_-_present?byZone=marvel_site_zone&offset=0&byType=comic_series&dateStart=&dateEnd=&orderBy=release_date+asc&byId=32866&limit=20&count=16'
amazing_spiderman_2018_url = 'https://www.marvel.com/comics/series/24396/the_amazing_spider-man_2018_-_present?byZone=marvel_site_zone&offset=20&byType=comic_series&dateStart=&dateEnd=&orderBy=release_date+asc&byId=24396&limit=108&count=20'
```

```python
driver.get(amazing_spiderman_2018_url)

action = ActionChains(driver)
more_btn = driver.find_element(By.CSS_SELECTOR,'.show-more')
action.move_to_element(more_btn).perform()
more_btn_small = driver.find_element(By.CSS_SELECTOR,'.btn-more-small')
action.move_to_element(more_btn_small).perform()
more_btn_small.click()
all_page_btn = driver.find_element(By.CSS_SELECTOR,'dl dd span:nth-child(2)')
action.move_to_element(all_page_btn).perform()
all_page_btn.click()

time.sleep(20)
```

```python
comic_items = driver.find_elements(By.CSS_SELECTOR,'.comic-item .row-item-image a img')
comic_url = driver.find_elements(By.CSS_SELECTOR,'.comic-item .row-item-image a')

title_list = []
img_list = []
id_list = []

for i in range(len(comic_items)):
    title = comic_items[i].get_attribute('title')
    img_src = comic_items[i].get_attribute('src')
    id = comic_url[i].get_attribute('href').split('/')[5]
    print(id)
    
    title_list.append(title)
    img_list.append(img_src)
    id_list.append(id)
    
comic_df = pd.DataFrame({
    'id':id_list,
    'title':title_list,
    'img_url':img_list
})
```

```python
comic_df.to_csv('data/amazing_spider_man_2018.csv')
```
